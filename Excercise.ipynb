# 13 ให้รับค่าน้ำหนัก และส่วนสูง แล้ว คำนวนค่า BMI จากสูตร BMI = น้ำหนักตัว(กก) หารด้วย ส่วนสูง(เมตร) ยกกำลังสอง
           
           สำหรับค่านี้กำหนดโดยองค์การอนามัยโลก (WHO)
 
           ค่าที่ได้น้อยกว่า 18.5 >>>> คุณผอมเกินไป

           ค่าที่ได้อยู่ระหว่าง 18.5-24.9 >>> คุณอยู่เกณฑ์เหมาะสม น้ำหนักตัวปกติ

           ค่าที่ได้อยู่ระหว่าง 25-29.9 >>> คุณน้ำหนักเกิน แต่ยังไม่เรียกว่าอ้วน

           ค่าที่ได้อยู่ระหว่าง 30-39.9 >>> คุณอ้วนแล้ว !

           ค่าที่ได้มากกว่าหรือเท่ากับ 40 >>> คุณอ้วนเกินไป อันตรายมาก !!!

```python
weight = float(input("please enter your weight in kg :"))
height = float(input("Please enter your height in cm :"))
heightm = height / 100
BMI = weight / (heightm **2)
if BMI < 18.5 :
    print("คุณผอมเกินไป")
elif BMI <= 24.9:
    print("คุณอยู่เกณฑ์เหมาะสม น้ำหนักตัวปกติ")
elif BMI <= 29.9:
    print("คุณน้ำหนักเกิน แต่ยังไม่เรียกว่าอ้วน")
elif BMI <= 39.9:
    print("คุณอ้วนแล้ว !")
else :
    print("คุณอ้วนเกินไป อันตรายมาก !!!")
```

# 21 Build factorial function (n! = n(n-1)(n-2)...1  for example 3! = 3 x 2 x 1 = 6)

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    return fact

c = factorial(6)
print("factorial =",c)
```

# 30 Build Contract Name and Number to Dictionary and have add delete lookup quit 

```python
contract = {}
running = True
while running:
    command = input('A for add , D for delete , L for lookup , Q for quit')
    if command == 'A' or command == 'a':
        name = input("Please enter your name :")
        phone = input('Please enter your number :')
        contract[name] = phone
    elif command == 'D' or command == 'd':
        name = input('Please enter name to delete')
        del contract[name]
    elif command == 'L' or command == 'l':
        name = input('Please enter name to lookup')
        print(contract[name])
    elif command == 'Q' or command == 'q':
        running = False
    else :
        print('Please enter valid character')
        
```

# 35 ผ่อนค่ารถ ดาวน์ 25% ผ่อน 5 ปี ดอกเบี้ย 3.3% ต่อปี ถ้าผ่อน 6 ปี ดอกเบี้ย 3.6% ต่อปี จงหาเงินที่ผ่อนต่อเดือน และเงินที่ต้องจ่ายทั้งหมด
ตัวอย่าง : ราคารถ 626,000 ดาวน์  (25%) 156,500 =  469,500<br><br>

ยอดจัดไฟแนนซ์ = 626,000-156,500 = 469,500.00<br><br>

   เงินต้น 469,500 + ดอกเบี้ย 77,467.50 = 546,967.50.-<br>

# 37 Chatbot

```python
import random
converse = True
greetings = ['hola','hello','hi','hey','Hello','Hi']
questions = ['How are you?','How are you doing?']
responses = ['Okay','I am fine']
validations = ['yes','yeah','ya','no','No','Nah','nah','yeaph']
verifications = ['Are you sure?','You sure?','you sure?','sure?','Sure?']
while converse:
    userinput = input('please talk to me :')
    if userinput in greetings:
        random_greeting = random.choice(greetings)
        print(random_greeting)
    elif userinput in questions:
        random_response = random.choice(responses)
        print(random_response)
    elif userinput in verifications :
        random_verification = random.choice(validations)
        print(random_verification)
    elif 'sure' in userinput:
        response = 'Sure about what ?'
        print(response)
    else :
        print('I do not know what you said ! Good bye')
        converse = False
```
